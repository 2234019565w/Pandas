```python
#导入需要用的包
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#正常显示中文
plt.rcParams["font.sans-serif"] = ['SimHei']   
plt.rcParams['axes.unicode_minus'] = False
```

```python
#获取数据集
iris = pd.read_csv('iris.csv')
```

```python
#查看数据集
iris.head()
```

第一幅图

```python
X1 = iris['PetalLength']
X2 = iris['PetalWidth']
```

```python
y = iris['Name']
```

```python
y.unique()
```

```python
plt.figure(dpi=100)
plt.scatter(X1[y=='Iris-setosa'],     X2[y=='Iris-setosa'],     color='r',marker='o',label='Iris-setosa') 
plt.scatter(X1[y=='Iris-versicolor'],X2[y=='Iris-versicolor'],  color='g',marker='o',label = 'Iris-Versicolour')    
plt.scatter(X1[y=='Iris-virginica'],  X2[y=='Iris-virginica'],  color='b',marker='o',label='Iris-Virginica')
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.legend(loc='upper left')
plt.title('PetalLength和PetalWidth之间的散点图')
plt.show()
```

第二幅图

```python
X3 = iris['SepalLength']
X4 = iris['SepalWidth']
```

```python
plt.figure(dpi=100)
plt.scatter(X3[y=='Iris-setosa'],     X4[y=='Iris-setosa'],     color='c',marker='o',label='Iris-setosa') 
plt.scatter(X3[y=='Iris-versicolor'], X4[y=='Iris-versicolor'], color='y',marker='o',label = 'Iris-Versicolour')    
plt.scatter(X3[y=='Iris-virginica'],  X4[y=='Iris-virginica'],  color='m',marker='o',label='Iris-Virginica')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend(loc='upper left')
plt.title('SepalLength和SepalWidth之间的散点图')
plt.show()
```
